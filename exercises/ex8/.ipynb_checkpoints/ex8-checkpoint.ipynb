### Estimating parameters for a Gaussian Distribution


To perform anomaly detection, you will first need to fit a model to the data’s
distribution.

`estimateGaussian` takes as input the data matrix X and should output an n-dimension vector
mu that holds the mean of all the n features and another n-dimension vector
sigma2 that holds the variances of all the features.

Visualize the contours of the fitted Gaussian distribution.

```python
def estimateGaussian(x):
    """
    Returns mean and variance (sigma2) by feature.
    """
    # means
    mu = (1 / len(x)) * np.sum(x, axis=0)

    # co-variance
    Sigma = (1 / len(x)) * (x - mu).T.dot((x - mu))

    return mu, Sigma
```

```python
def predictGaussian(x, mu, Sigma):
    """
    Returns pvals of x given a Gaussian model mean and govariance mu and Sigma. 
    """
    
    # create gaussian model
    model = sp.stats.multivariate_normal(mean=mu, cov=Sigma)

    # predict with x_val
    pval = model.pdf(x)

    return pval
```

```python
# visualize estimated Gaussian as contour plot

# compute stats
mu, sigma2 = estimateGaussian(X1)

# scatter
plt.scatter(X1[:, 0], X1[:, 1], marker="x", c=predictGaussian(X1, mu, sigma2))
plt.xlabel("Latency (ms)")
plt.ylabel("Throughput (mb/s)")

# contour
# prepare mesh for contour plot
x, y = np.mgrid[0:30:1, 0:30:1]
pos = np.empty(x.shape + (2,))
pos[:, :, 0] = x
pos[:, :, 1] = y
rv = sp.stats.multivariate_normal(mean=mu, cov=sigma2)
plt.contour(x, y, rv.pdf(pos), alpha=0.5)
plt.show()
```

### Selecting the threshold

Now that you have estimated the Gaussian parameters, you can investigate
which examples have a very high probability given this distribution and which
examples have a very low probability. The low probability examples are
more likely to be the anomalies in our dataset. One way to determine which
examples are anomalies is to select a threshold based on a cross validation
set. In this part of the exercise, you will implement an algorithm to select
the threshold ε using the F1 score on a cross validation set.

In `selectThreshold`, use a cross validation set, where label y=1 corresponds to anomalous example, and y=0 corresponds to a normal example. The function should return 2 values: selected threshold $\epsilon$ and corresponding F1 score:

$$F_1 = \frac{2 \cdot precision \cdot recall}{ precision + recall}$$

```python
def selectThreshold(pval,y_val):
    """
    Returns optimal probability threshold to select for anomalous data points.
    """
    # thresholds to check
    thresh_vals = np.linspace(1.01*min(pval), max(pval), 1000)

    # init variables
    bestEpsilon = 0
    bestF1 = 0
    f1 = 0
    
    for i, thresh in enumerate(thresh_vals):

        # Given threshold, compute F1 score
        predictions = pval < thresh

        tp = np.sum((predictions == y_val) & (y_val == 1))
        fp = np.sum((predictions == 1) & (y_val == 0))
        fn = np.sum((predictions == 0) & (y_val == 1))
        
        prec = tp / (tp + fp)
        reca = tp / (tp + fn)

        epsilon = thresh
        f1 = (2 * prec * reca) / (prec + reca)

        if f1 > bestF1:
            bestF1 = f1
            bestEpsilon = epsilon

    return {"epsilon": bestEpsilon, "f1": bestF1}
```

```python
pval = predictGaussian(X1val, mu, sigma2)
thresh = selectThreshold(pval,y1val)
print(thresh)
```

```python
# visualize
pval = predictGaussian(X1, mu, sigma2)

plt.scatter(
    X1[:, 0], X1[:, 1], c= pval <= thresh['epsilon'], cmap=plt.cm.cool, marker="x"
)
plt.title('Cross-validation set')
plt.xlabel("Latency (ms)")
plt.ylabel("Throughput (mb/s)")

plt.tight_layout()
plt.show()
```

### Movie Ratings dataset

The matrix Y (a num movies × num users matrix) stores the ratings y
(i,j)
(from 1 to 5). The matrix R is an binary-valued indicator matrix, where
R(i, j) = 1 if user j gave a rating to movie i, and R(i, j) = 0 otherwise. The
objective of collaborative filtering is to predict movie ratings for the movies
that users have not yet rated, that is, the entries with R(i, j) = 0. This will
allow us to recommend the movies with the highest predicted ratings to the
user.
To help you understand the matrix Y, the script ex8 cofi.m will compute
the average movie rating for the first movie (Toy Story) and output the
average rating to the screen.

### Collaborative filtering learning algorithm

This algorithm predicts the rating for movie $i$ by user $j$ using the parameter vectors.

```python
# Collaborative filtering cost function:


def cofiCostFunc(params, y, R, num_users, num_movies, num_features, lbda=0.0):
    """
    Returns value of regularized cost function.
    """
    # reshape params
    x = params[: (num_movies * num_features)].reshape(num_movies, num_features)
    Theta = params[num_movies * num_features :].reshape(num_users, num_features)

    # compute cost
    J = (1 / 2) * np.sum(((x.dot(Theta.T) - y) ** 2) * R)

    return J


# Collaborative filtering gradient:

def cofiGrad(params, y, R, num_users, num_movies, num_features, lbda=0.0):
    """
    Returns value of regularized gradients for theta and X.
    """
    # reshape params
    x = params[: (num_movies * num_features)].reshape(num_movies, num_features)
    Theta = params[num_movies * num_features :].reshape(num_users, num_features)

     
    X_grad = np.sum( (((x.dot(Theta.T) - y)) * R).dot(Theta) , axis = 1)
    
    Theta_grad = np.sum( (((x.dot(Theta.T) - y)) * R).T.dot(x) , axis = 1)
    
    return np.concatenate([X_grad, Theta_grad])

def fitModel(params, y, R, num_users, num_movies, num_features, lbda=0.0):
    """
    Fits model parameters to data according to our hypothesis,
    cost and gradient functions.
    """
    opt_weights = sp.optimize.minimize(
        fun=cofiCostFunc,tnc: invalid gradient vector from minimized function.
        x0=params,
        args=(y, R, num_users, num_movies, num_features, lbda),
        method="TNC",
        jac=cofiGrad,
    )
    return opt_weights
```
